```python
# build the lightgbm model
import lightgbm as lgb

clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)
```

```python
# predict the results USING LIGHTGBM
y_pred=clf.predict(X_test)
```

```python
# view accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred, y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
from catboost import Pool, CatBoostClassifier, cv, CatBoostRegressor
```

```python
cate_features_index = ['form_field47']
cate_features_index
```

```python
#let us make the catboost model, use_best_model params will make the model prevent overfitting
model = CatBoostClassifier(iterations=3500, learning_rate=0.07, l2_leaf_reg=3.5, depth=8, rsm=0.98, loss_function= 'Logloss', eval_metric='AUC',use_best_model=True,random_seed=42)
```

```python
#now just to make the model to fit the data
model.fit(X_train, y_train,cat_features=cate_features_index,eval_set=(X_test,y_test))
```

```python
# predict the results
y_pred=model.predict(X_test)
```

```python
# view accuracy
from sklearn.metrics import accuracy_score
y_test = y_test.astype(float)
accuracy=accuracy_score(y_pred, y_test)
print('CATBOOST Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred)))
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```
